```python
# Lag the features
lagged_df = df.copy()
lagged_df.iloc[:, 1:] = df.iloc[:, 1:].shift(1)  # Lag all columns except datetime
```

```python
# Drop the first row with NaN values
lagged_df.dropna(inplace=True)
```

```python
# Max and Min values of the return
max_return = lagged_df['returns'].max()
min_return = lagged_df['returns'].min()

max_return, min_return
```

```python
# Extract features and target variable
X = lagged_df.drop(['returns', 'datetime', 'o'], axis=1)  
y = lagged_df['returns']
```

```python
# Split the data into training and testing set for Time series data
train_size = int(0.8 * X.shape[0])
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
X_train.shape, X_test.shape
```

```python
# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Check the shape of your input data
print(X_train_scaled.shape)

# If the input data is 2D, add a dimension to make it 3D
if len(X_train_scaled.shape) == 2:
    X_train_scaled = np.expand_dims(X_train_scaled, axis=1)

# Now the shape of X_train_scaled should be (batch_size, 1, features)
print(X_train_scaled.shape)
```

```python
# Check the shape of your input data
print(X_test_scaled.shape)

# If the input data is 2D, add a dimension to make it 3D
if len(X_test_scaled.shape) == 2:
    X_test_scaled = np.expand_dims(X_test_scaled, axis=1)

print(X_test_scaled.shape)
```

```python
# Getting number of positive and negative returns in the training set
pos_returns = (y_train > 0).sum()
neg_returns = (y_train < 0).sum()

pos_returns, neg_returns
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
from kerastuner.tuners import BayesianOptimization

# Define the function to build the GRU model
def build_gru_model(hp):
    model = Sequential()
    model.add(GRU(units=128,
                  activation='relu',
                  input_shape=(X_train_scaled.shape[1], X_train_scaled.shape[2])))
    model.add(Dropout(rate=hp.Float('dropout_rate', min_value=0.0, max_value=0.1, step=0.1)))
    model.add(Dense(1))  # Output layer for regression task

    model.compile(optimizer='adam',
                  loss='mse',
                  metrics=['mse'])
    return model

# Define the BayesianOptimization tuner
tuner = BayesianOptimization(
    build_gru_model,
    objective='val_loss',  # Choose the metric to optimize (e.g., validation loss)
    max_trials=10,  # Total number of trials (hyperparameter combinations) to test
    executions_per_trial=1,  # Number of executions per trial (training runs)
    directory='bayesian_opt',  # Directory to store the tuning logs and checkpoints
    project_name='gru_hyperparams'  # Name for the tuning project
)

# Start the hyperparameter search
tuner.search(X_train_scaled, y_train,
             epochs=100,  # Number of epochs to train the model
             validation_data=(X_test_scaled, y_test))  # Validation data for monitoring performance

# Get the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
best_units = best_hps.get('units')
best_dropout_rate = best_hps.get('dropout_rate')

# Build the GRU model with the best hyperparameters
best_gru_model = build_gru_model(best_hps)

# Train the model with the best hyperparameters and early stopping
best_gru_model.fit(X_train_scaled, y_train,
                    epochs=100,  
                    validation_data=(X_test_scaled, y_test))  # Validation data for monitoring performance
```

```python
# Saving the model
best_gru_model.save('gru_model.h5')
```

```python
# Printing best hyperparameters
print(f"Best units: {best_units}")
print(f"Best dropout rate: {best_dropout_rate}")

# Evaluate the model
train_loss = best_gru_model.evaluate(X_train_scaled, y_train)
test_loss = best_gru_model.evaluate(X_test_scaled, y_test)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from scipy.optimize import dual_annealing

# Define the SVR model
kernel = 'rbf'
C = 1   
gamma = 1
epsilon = 0.1
svr_model = SVR(kernel=kernel, C=C, gamma=gamma, epsilon=epsilon)
```

```python
# Define the time series version of the SVR model
from sklearn.svm import SVR
class TimeSeriesSVR:
    def __init__(self, C, epsilon, gamma, lag):
        self.C = C
        self.epsilon = epsilon
        self.gamma = gamma
        self.lag = lag
        self.model = SVR(C=self.C, epsilon=self.epsilon, gamma=self.gamma)

    def prepare_data(self, X, y):
        X_lagged = X.shift(self.lag).dropna()
        y_lagged = y.shift(self.lag).dropna()
        X_lagged = X_lagged.values
        y_lagged = y_lagged.values
        return X_lagged, y_lagged

    def fit(self, X, y):
        X_lagged, y_lagged = self.prepare_data(X, y)
        self.model.fit(X_lagged, y_lagged)

    def predict(self, X):
        X_lagged = X.shift(self.lag).dropna()
        X_lagged = X_lagged.values
        return self.model.predict(X_lagged)

    def score(self, X, y):
        X_lagged, y_lagged = self.prepare_data(X, y)
        return self.model.score(X_lagged, y_lagged)

    def mse(self, X, y):
        y_pred = self.predict(X)
        return mean_squared_error(y.shift(self.lag).dropna(), y_pred)

    def objective(self, hyperparameters):
        self.C, self.epsilon, self.gamma = hyperparameters
        self.model = SVR(C=self.C, epsilon=self.epsilon, gamma=self.gamma)
        return self.mse(X_train, y_train)

    def optimize(self):
        bounds = [(0.1, 100), (0.01, 1), (0.1, 1)]
        result = dual_annealing(self.objective, bounds)
        return result.x
    
# Instantiate the model
ts_svr = TimeSeriesSVR(C=1, epsilon=0.1, gamma='scale', lag=1)

# Fit the model
ts_svr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
# Make predictions
y_pred = ts_svr.predict(X_test_scaled)

# Evaluate the model
mse = ts_svr.mse(X_test_scaled, y_test)
mse
```
